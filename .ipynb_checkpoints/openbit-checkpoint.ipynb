```python
# -*- coding:utf-8 -*-
import poloniex
import time
import json

# poloniex API
polo = poloniex.Poloniex()
# 2017/12/12 AM 12:00 @JP
unix_time = 1514646000

#4h間隔で過去180日分 UNIX時間で取得 
chart_data = polo.returnChartData('USDT_BTC', period=14400, start=unix_time-polo.DAY*180, end=unix_time)

```

```python
import pandas as pd

# DataFrameに変換
df = pd.DataFrame(chart_data)
df.tail(10)
```

```python
# 短期線：窓幅1日
data_s = df['close'].rolling(window=6).mean() 
 
# 長期線：窓幅5日
data_l = df['close'].rolling(window=6*5).mean()
```

```python
import matplotlib.pyplot as plt
# 描画を綺麗に表示する
from matplotlib.pylab import rcParams
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

rcParams['figure.figsize'] = 15, 6

# 短期線と長期線もプロット
plt.plot(df['close'], color='#7f8c8d')
plt.plot(data_s, color='#f1c40f')  # 短期線
plt.plot(data_l, color='#2980b9')  # 長期線
plt.show()
```

もとの価格だと学習が進まなかったため、最小、最大値で正規化
- TODO fit_transformがよく分かってない

```python
import numpy as np


data = pd.DataFrame([float(i.get('close')) for i in chart_data])
mss = MinMaxScaler()
data = pd.DataFrame(mss.fit_transform(data))
data = data[0].astype(np.float32)

# データを入力変数xと出力変数tに切り分け
x, t = [], []
N = len(data)
M = 30 # 入力変数の数：直近30サンプルを使用
for n in range(M, N):
 # 入力変数と出力変数の切り分け
 _x = data[n-M: n] # 入力変数
 _t = data[n] # 出力変数
 # 計算用のリスト(x, t)に追加していく
 x.append(_x)
 t.append(_t)

# numpyに変換
# x.shape : 1050,30 t.shape : 1050,1
x = np.array(x)
x = x.reshape(x.shape[0], x.shape[1], 1)
t = np.array(t).reshape(len(t), 1) 
```

- TODO keras LSTMの2つのdropoutの意味

```python
# model
from keras.models import Sequential
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping

length_of_sequence = x.shape[1] # 30 = M
in_out_neurons = 1
# n_hidden: 出力次元
n_hidden = 300
model = Sequential()
# return_sequenceとは、TrueにしてRNNレイヤーの毎時刻の出力を得るか、Falseにして最後の時刻のみの出力を得るか、のフラグ
# batch_iuput_shape: ([バッチサイズ，step数，特徴の次元数])
model.add(LSTM(units=n_hidden,
               batch_input_shape=(None, length_of_sequence, in_out_neurons), 
               return_sequences=False
              ))
model.add(Dropout(0.5))
model.add(Dense(in_out_neurons))
model.add(Activation("linear"))
optimizer = Adam(lr=0.001)
model.compile(loss="mean_squared_error", optimizer=optimizer)

model.summary()
```

```python
early_stopping = EarlyStopping(monitor='val_loss', mode='auto', patience=20)
model.fit(x, t,
          batch_size=20,
          epochs=30,
          validation_split=0.1,
          shuffle=True,
          callbacks=[early_stopping]
          )
```

```python
predicted = model.predict(x)
predicted
```

```python
plt.figure()
plt.plot(range(0, len(predicted)), predicted, color="r", label="predict_data")
plt.plot(range(0, len(t)), t.reshape(1050), color="b", label="row_data")
plt.legend()
plt.show()
```
